```python
from tqdm.notebook import tqdm
import pandas as pd
```

```python


df = pd.read_csv('./Questões/Questões - Dataset Unicamp IR - Página1.csv')
df['id'] = df.index
df = df[['id','Pergunta']]
df.to_csv('./Questões/portuguese_general_queries.tsv',index=False)
```

```python
df = pd.read_csv('./Questões/portuguese_general_queries.tsv')
df
```

```python
def trec_run(save_path, topics, top_k, searcher):
  os.makedirs('./runs',exist_ok=True)
  with open(save_path, 'w') as fout:
    for id, query in tqdm(topics):
      hits = searcher.search(query, top_k)
      for idx, hit in enumerate(hits):
        hit_raw =  ast.literal_eval(hit.raw)
        fout.write(f"{id}\tQ0\t{hit.docid}\t{idx+1}\t{hit.score}\t{hit_raw}\tBM25\n")

analyzer = get_lucene_analyzer("pt")
index_path = 'bm25/passage_v2'  # Caminho para o índice pré-construído
searcher = LuceneSearcher(index_path)
searcher.set_analyzer(analyzer)

topics = list(df.itertuples(index=False, name=None))
trec_run('./runs/portuguese_general_queries_passages_v2.tsv', topics, 1000, searcher)
```
