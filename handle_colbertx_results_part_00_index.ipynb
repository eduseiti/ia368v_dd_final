```python
result_files = glob.glob("validation_queries/colbertx_retrieval/ranking_index_0*.tsv")
```

```python
result_files
```

```python
qrels_dfs = []

for result_file in result_files:
    qrels_dfs.append(pd.read_csv(result_file, sep='\t', header=None, names=['query_id', 'colbert_id', 'ranking', 'score']))
```

```python
qrels_dfs[0].shape
```

```python
qrels_df = pd.concat(qrels_dfs)
```

```python
for group_name, group_df in qrels_df.groupby('query_id'):
    
    print(queries_df[queries_df['query_id'] == group_name]['query_text'].to_numpy()[0])
    
    display(group_df.sort_values(['score'], ascending=False)[:10].merge(corpus_df))
```
