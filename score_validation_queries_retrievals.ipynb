## Analyze queries documents

```python
results = []

for group_name, group_df in bm25_df.groupby('id'):

    print("query_id={}\n".format(group_name))

    bm25_passage_ids = group_df['passagem-id'].to_numpy()

    bm25_mT5_query_df = bm25_mT5_df[bm25_mT5_df['id'] == group_name]
    bm25_mT5_passage_ids = bm25_mT5_query_df['doc_id'].to_numpy()

    if group_name < 100:
        colbertx_query_df = colbertx_p1_df[colbertx_p1_df['query_id'] == group_name]
    else:
        colbertx_query_df = colbertx_p2_df[colbertx_p2_df['query_id'] == group_name]

    colbertx_passage_ids = colbertx_query_df['clueweb_id'].to_numpy()

    for passage_id in np.sort(np.union1d(np.union1d(bm25_passage_ids, bm25_mT5_passage_ids), colbertx_passage_ids[:10])):

        bm25_pos = np.where(bm25_passage_ids == passage_id)[0]
        bm25_mT5_pos = np.where(bm25_mT5_passage_ids == passage_id)[0]
        colbertx_pos = np.where(colbertx_passage_ids == passage_id)[0]

        if len(bm25_pos) > 0:
            passage_text = group_df.iloc[bm25_pos[0]]['passagem-by-BM25']
        elif len(bm25_mT5_pos) > 0:
            passage_text = bm25_mT5_query_df.iloc[bm25_mT5_pos[0]]['passage-by-bm25-mt5']
        else:
            passage_text = colbertx_query_df.iloc[colbertx_pos[0]]['passage']

        results.append({'query_id': group_name,
                        'clueweb_id': passage_id,
                        'bm25_pos': -1 if len(bm25_pos) == 0 else int(bm25_pos[0]),
                        'bm25_mT5_pos': -1 if len(bm25_mT5_pos) == 0 else int(bm25_mT5_pos[0]),
                        'colbertx_pos': -1 if len(colbertx_pos) == 0 else int(colbertx_pos[0]),
                        'query': group_df.iloc[0]['query'],
                        'passage': passage_text})
```

```python
results_df = pd.DataFrame(results)
```

```python
results_df = pd.read_csv(os.path.join(RETRIEVALS_FOLDER, "pipelines_retrievals_comparisson.tsv"), sep='\t')
```

```python
compiled_results = []

for group_name, group_df in results_df.groupby('query_id'):
    returned_docs_pos = group_df[['bm25_pos', 'bm25_mT5_pos', 'colbertx_pos']].to_numpy()

    missing_docs = np.zeros(returned_docs_pos.shape[0])
    missing_docs_all = np.zeros(returned_docs_pos.shape[0])

    for i in range(returned_docs_pos.shape[0]):
        # print(returned_docs_pos[i])

        greater_than_9 = np.where(returned_docs_pos[i] > 9)[0]

        excluding_greater_than_9 = returned_docs_pos[i].copy()
        excluding_greater_than_9[greater_than_9] = -1

        # print(excluding_greater_than_9)

        missing_pos_all = np.where(returned_docs_pos[i] == -1)[0]
        missing_pos = np.where(excluding_greater_than_9 == -1)[0]

        missing_docs_all[i] = len(missing_pos_all)
        missing_docs[i] = len(missing_pos)

    compiled_results.append({'query': group_name,
                             'single_pipeline': len(np.where(missing_docs_all == 2)[0]),
                             '2_pipelines': len(np.where(missing_docs_all == 1)[0]),
                             '3_pipelines': len(np.where(missing_docs_all == 0)[0]),
                             'single_pipeline_in_10': len(np.where(missing_docs == 2)[0]),
                             '2_pipelines_in_10': len(np.where(missing_docs == 1)[0]),
                             '3_pipelines_in_10': len(np.where(missing_docs == 0)[0]),
                             })
```

```python
compiled_results_df = pd.DataFrame(compiled_results)
```
